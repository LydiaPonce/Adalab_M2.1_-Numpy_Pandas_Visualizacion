```python
import pandas as pd
df_attack2 = pd.read_csv("./attack2.csv", index_col = 0)
df_histo = pd.read_csv("./attacks_historico.csv", index_col = 0)
df_adicional = pd.read_csv("./attacks_adicional.csv", index_col = 0)
```

2- **Chequeamos que los nombres de las columnas que sean iguales entre dataframes son iguales. ¿Están todas en minúsculas? ¿Tienen espacios? En caso de que no sean todas iguales unificad el nombre de las columnas.**

```python
columnas_nuevas = {col :  col.lower().replace(" ","_") for col in df_attack2.columns}
df_attack2.rename(columns = columnas_nuevas, inplace= True)
```

```python
#En attack2 están todas en minúsculas, en historico no.
#Los espacios han sido sustituidos por "_"
columnas_nuevas = {col :  col.lower().replace(" ","_") for col in df_histo.columns}
df_histo.rename(columns = columnas_nuevas, inplace= True)
df_histo.head(2)
```

3- **En este momento nos habremos dado cuenta que el dataframe que generamos en el pair programming Pandas III es similar al dataframe de los datos historicos. Si recordamos, eliminamos algunas columnas en la sesión III que no eran útiles, pero esas columnas están todavía en el histórico. Si no fuera por eso podríamos unir esos dataframes. Realizad los cambios que creais necesarios para que los dataframes sean iguales y luego juntadlos usando el mejor método.**

```python
df_histo.drop(["href","unnamed:_22","unnamed:_23", "case_number.1", "case_number.2", "original_order", "investigator_or_source", "pdf", "href_formula"], axis = 'columns', inplace=True)
del df_histo["href"]
```

```python
df_histo.set_index("case_number",iplace=True)
```

```python
del df_attack2["href"]
```

```python
df_attack2.head()
```

```python
df_histo.head()
```

```python
df_juntas = pd.concat([df_attack2, df_histo], axis=0, ignore_index = True)
df_juntas.head()
#No utilices el join porque te duplica las columnas
```

4- **Es el momento de unir la información de todos los ataques de tiburón que acabamos de crear con el dataframe de información adicional. Debatid que método es el más correcto para unir toda esta información y ejecutadlo.**

📌 NOTA Cuando hagáis la unión considerad que la información que más nos interesa es la del dataframe que creamos en el apartado anterior, lo cual determinará el tipo de unión.

```python
columnas_nuevas2 = {col :  col.lower().replace(" ","_") for col in df_adicional.columns}
df_adicional.rename(columns = columnas_nuevas2, inplace= True)
df_adicional.head()
```

```python
mergeado_izq = pd.merge(left=df_juntas, right=df_adicional, how='left',
                        left_on='case_number', right_on='case_number')

mergeado_izq.head()
```

```python
mergeado_izq.to_csv("attackito_supremo.csv")
```
